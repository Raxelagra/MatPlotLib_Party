```python
#My exploration NO.3 : Line sub-plots
#set font dictionary to improve labels
font1 = {'family':'serif','color':'navy','size':10}
font2 = {'family':'serif','color':'darkred','size':15}
font3 = {'family':'serif','color':'k','size':25}

#adjust size of subplots
plt.subplots(2, 2, figsize=(6,6))

#since plots are to be on individual graphs, each plt will be treated as a subplot
#plot 1
x11=np.array([0,1,2,3])
y11=np.array([3,8,1,10])

plt.subplot(3,2,1) #line subplot for a group containing 3 rows and 2 columns at postion 1
plt.plot(x11,y11, marker = '*',markersize = 15, color = 'magenta', linestyle = '--')
#improve fonts on labels and title
plt.title('Primary 1', fontdict = font2)
plt.xlabel('x-axis', fontdict = font1)
plt.ylabel('y-axis', fontdict = font1)
plt.legend(['Primary 1']) #set legend

#plot subsequent subplots in same fashion as plot 1 but vary the colours at each stage
#plot2
x22=np.array([0,1,2,3])
y22=np.array([10,20,30,40])

plt.subplot(3,2,2) #line subplot for a group containing 3 rows and 2 columns at postion 2
plt.plot(x22,y22,marker = '*',markersize = 15, color = 'navy', linestyle = '--')
plt.title('Primary 2', fontdict = font2)
plt.xlabel('x-axis', fontdict = font1)
plt.ylabel('y-axis', fontdict = font1)
plt.legend(['Primary 2'])


#plot3
x33=np.array([0,1,2,3])
y33=np.array([10,20,30,40])

plt.subplot(3,2,3) #line subplot for a group containing 3 rows and 2 columns at postion 3
plt.plot(x33,y33,marker = '*',markersize = 15, color = 'darkgreen', linestyle = '--')
plt.title('Primary 3', fontdict = font2)
plt.xlabel('x-axis', fontdict = font1)
plt.ylabel('y-axis', fontdict = font1)
plt.legend(['Primary 3'])


#plot4
x44=np.array([0,1,2,3])
y44=np.array([10,20,30,40])

plt.subplot(3,2,4) #line subplot for a group containing 3 rows and 2 columns at postion 4
plt.plot(x44,y44,marker = '*',markersize = 15, color = 'crimson', linestyle = '--')
plt.title('Primary 4', fontdict = font2)
plt.xlabel('x-axis', fontdict = font1)
plt.ylabel('y-axis', fontdict = font1)
plt.legend(['Primary 4'])


#plot5
x55=np.array([0,1,2,3])
y55=np.array([10,20,30,40])

plt.subplot(3,2,5) #line subplot for a group containing 3 rows and 2 columns at postion 5
plt.plot(x55,y55,marker = '*',markersize = 15, color = 'navy', linestyle = '--')
plt.title('Primary 5', fontdict = font2)
plt.xlabel('x-axis', fontdict = font1)
plt.ylabel('y-axis', fontdict = font1)
plt.legend(['Primary 5'])


#plot6
x66=np.array([0,1,2,3])
y66=np.array([10,20,30,40])

plt.subplot(3,2,6) #line subplot for a group containing 3 rows and 2 columns at postion 6
plt.plot(x66,y66,marker = '*',markersize = 15, color = 'darkgreen', linestyle = '--')
plt.title('Primary 6', fontdict = font2)
plt.xlabel('x-axis', fontdict = font1)
plt.ylabel('y-axis', fontdict = font1)
plt.legend(['Primary 6'])

#adjust position of the subplots to prevent overlaps and improve the look of the output
plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=1.2, 
                    top=0.8, 
                    wspace=0.8, 
                    hspace=1.4)

#set Super Title for the total plot and apply font improvement
plt.suptitle('Trend in Performance in the Primary School of SCHOLARS PREPARATORY SCHOOL in Year 2021', fontdict = font3)
plt.show()
```
